```python
df_birth = pd.read_csv('https://raw.githubusercontent.com/kryxsofiaengay/TopicProposal_TeamFighters/main/Datasets1/adolescentBirthRate.csv')
df_birth = df_birth.drop('Indicator', axis = 1)
df_birth.columns = ['Location', 'Period', 'Adolescent birth rate per 1000 women aged 15-19 years']
df_birth.head()
```

<p align="justify">&nbsp;&nbsp;&nbsp;&nbsp;When compared to a list of ten randomly chosen nations, including developing and industrialized nations, Egypt still has the highest incidence of adolescent pregnancies, indicating that most African nations, including the more developed ones, still have high rates of teen pregnancies.Up to 38% of all female homicides worldwide are the result of romantic relationships. In addition to intimate relationship violence, 6% of women worldwide claim to have experienced non-partner sexual assault, while data on non-partner sexual violence are scarcer. The majority of intimate relationship and sexual violence against women is committed by men.Women are now more likely to be exposed to abusive relationships and recognized risk factors due to lockdowns caused by the COVID-19 epidemic and its social and economic effects, while also having less access to resources. Humanitarian crises and displacement situations may intensify pre-existing violence against women, such as that committed by intimate partners and non-partner sexual assault, and they may also give rise to new types of violence against women.

```python
random_count = df_birth[df_birth['Period'] == 2018]
random_count = random_count.sample(n = 10, random_state=11)
random_count = pd.concat([random_count, df_birth[(df_birth['Period'] == 2018) & (df_birth['Location'] == 'Egypt')]])
```

```python
plt.figure(figsize=(10,9), dpi = 200)
plt.title('Teenage Pregnancy Rates in 2018')
fig = sns.barplot(data = random_count, x = 'Location', y = 'Adolescent birth rate per 1000 women aged 15-19 years');
fig.set_ylabel('Adolescent birth rate per 1000 women aged 15-19 years', fontsize=12);
fig.set_xlabel('Location', fontsize=12);
plt.tick_params(labelsize=10);
```

 <p align="justify">&nbsp;&nbsp;&nbsp;&nbsp; Afghanistan's adolescent fertility rate (births per 1,000 women aged 15 to 19), as published by the World Bank's collection of development indicators, was estimated to be 87.0%. Actual statistics, historical data, estimates, and predictions for Afghanistan's adolescent fertility rate (births per 1,000 women aged 15 to 19) were obtained from the World Bank.

```python
egypt_birth = df_birth[(df_birth['Period'] == 2018) & (df_birth['Location'] == 'Egypt')]['Adolescent birth rate per 1000 women aged 15-19 years'].values

plt.figure(figsize = (10, 9))
fig = sns.boxplot(data = df_birth[df_birth['Period'] == 2018], y = 'Adolescent birth rate per 1000 women aged 15-19 years')
fig.axhline(egypt_birth, linestyle = '--', color = 'r',  )
fig.set_ylabel('Adolescent birth rate per 1000 women aged 15-19 years', fontsize=15);
plt.tick_params(labelsize=15);
```

Egypt is shown as a red dashed line.

<p align="justify">&nbsp;&nbsp;&nbsp;&nbsp; This boxplot of adolescent pregnancies in 2018 shows that Egypt is above the third quantile, indicating that it is still performing poorly relative to the majority of nations from that year that are not in Africa.

```python
df_sanit = pd.read_csv('https://raw.githubusercontent.com/kryxsofiaengay/TopicProposal_TeamFighters/main/Datasets1/atLeastBasicSanitizationServices%20(1).csv')
df_sanit = df_sanit.drop('Indicator', axis = 1)
df_sanit.columns = ['Location', 'Period', 'Area', 'Population using Basic Sanitation Services (%)']
df_sanit.head()
```

```python
plt.figure(figsize = (11, 8), dpi = 100)
plt.title('Percentage of Rural/Urban/Total population in India with Basic Sanitation Services')
sns.barplot(data = df_sanit[df_sanit['Location'] == 'India'], x = 'Period', y = 'Population using Basic Sanitation Services (%)', hue = 'Area', hue_order=['Rural', 'Urban', 'Total'])
```

<p align="justify">&nbsp;&nbsp;&nbsp;&nbsp; When it comes to important sanitization, it appears that India's rural and urban populations differ significantly. While both have been improving at a linear pace, with the rural rate being greater, the urban population's sanitization is unmistakably much better, with roughly 72% of individuals having at least basic sanitization in 2017. In contrast, the rural population's sanitization only just above 50% in 2017.

<p align="justify">&nbsp;&nbsp;&nbsp;&nbsp; This truly clarifies us more on what's genuinely taking place in India. Only 50% of the population in India has access to even the most basic hand sanitization facilities since 72.18% of the country's population lives in rural areas (urban areas make up the remainder). The Total (Urban and Rural) does in fact take this into account, but it does not reflect the fact that the majority of the population is faring worse.
